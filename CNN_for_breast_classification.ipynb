```python
import numpy as np
import tensorflow as tf


from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
training_set = train_datagen.flow_from_directory('/content/drive/MyDrive/BreastDataset',
                                                 target_size = (300, 300),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
from keras.applications import VGG16
from keras.layers import Dense, Input, Flatten
from keras.models import Model
import numpy as np 
import tensorflow as tf



model_vgg16_conv = VGG16(include_top=False)
input_ = tf.keras.layers.Input(shape=(300,300,3),name = 'image_input')
output_vgg16_conv = model_vgg16_conv(input_)
#Add the fully-connected layers 
x = Flatten(name='flatten')(output_vgg16_conv)
x = Dense(128, activation='relu', name='fc1')(x)
x = Dense(64, activation='relu', name='fc2')(x)
x = Dense(4, activation='softmax', name='predictions')(x)


breast_classifier = Model(input_,x)

breast_classifier.compile(optimizer = 'adam', 
            loss = 'categorical_crossentropy', 
            metrics = ['accuracy'])
```

```python
import tensorflow as tf
breast_classifier = tf.keras.models.load_model('drive/MyDrive/Colab/breast_classifier_less_dense_ht_sch.h5')
```

```python
breast_classifier.summary()
```

```python
from datetime import date
from datetime import datetime
import pickle
for x in range (0,100):
  history = breast_classifier.fit(x=training_set,epochs=1)
  breast_classifier.save('drive/MyDrive/Colab/breast_classifier_less_dense_ht.h5')
  now = datetime.now()
  pickle.dump(history.history, open( "drive/MyDrive/Colab/history_for_less_dense%s.p"%str(now), "wb" ) )
```
